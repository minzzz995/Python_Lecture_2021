```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target,
random_state=42)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print(tree.score(X_train, y_train))
print(tree.score(X_test, y_test))
```

```python
tree = DecisionTreeClassifier(max_depth=4,random_state=0)
tree.fit(X_train, y_train)
print(tree.score(X_train, y_train))
print(tree.score(X_test, y_test))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file="tree.dot", class_names=["악성","양성"],
feature_names=cancer.feature_names,impurity=False, filled=True)
```

```python
import pandas as pd
ram_prices = pd.read_csv('c:/mysql/ram_price.csv')
plt.semilogy(ram_prices.date,ram_prices.price)
plt.show()
```

```python
plt.plot(ram_prices.date, ram_prices.price)
```

```python
from sklearn.tree import DecisionTreeRegressor
data_train = ram_prices[ram_prices.date<2000]
data_test = ram_prices[ram_prices.date>=2000]
```

```python
data_train.shape
```

```python
data_test.shape
```

```python
X_train = data_train.date.to_numpy()[:,np.newaxis]
X_train
```

```python
y_train = np.log(data_train.price)
```

```python
dt = DecisionTreeRegressor().fit(X_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train, y_train)
```

```python
X_all = ram_prices.date.to_numpy()[:,np.newaxis]
```

```python
pred_dt = dt.predict(X_all)
pred_lr = lr.predict(X_all)
```

```python
price_dt = np.exp(pred_dt)
price_lr = np.exp(pred_lr)
```

```python
plt.semilogy(data_train.date, data_train.price, label="train data")
plt.semilogy(data_test.date, data_test.price, label="test data")
plt.semilogy(ram_prices.date, price_dt, label="pred of dt")
plt.semilogy(ram_prices.date, price_lr, label="pred of lr")
plt.legend()
plt.show()
```
