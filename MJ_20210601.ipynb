```python
url = "https://en.wikipedia.org/w/index.php?title=List_of_Game_of_Thrones_episodes&oldid=802553687"
r = requests.get(url)
r.text
html_soup = BeautifulSoup(r.text,"html.parser")
found_h1 = html_soup.find("h1") #페이지소스보기 -> ctrl+F하고 찾기
found_h1
```

```python
# 많은 contents를 찾을땐 h1처럼 하지 않고 반복문을 사용해서 찾는다
for found in html_soup.find_all("h2"): # h2레벨을 모두 찾겠다(19개)
    print(found.text) # 한 개당 2개씩 갯수를 포함하므로 8개 
```

```python
episodes = []
ep_tables = html_soup.find_all('table', 
                               class_ = "wikitable plainrowheaders wikiepisodetable") #표를 다 찾아보기
```

```python
for table in ep_tables:
    headers = [] #표의 header공간에 담을 공간 만들어주기
    rows = table.find_all('tr') #모든 rows를 다 찾아봐라
    for header in table.find("tr").find_all("th"): #"th"를 찾아서 리스트에 담아봐라
        headers.append(header.text)
        
    for row in table.find_all("tr")[1:]:
        values = [] #values가 담을 공간 만들어주기
        for col in row.find_all(["th","td"]): #columns에 나와있는 것을 하나씩 values에 넣어줘라
            values.append(col.text)
        if values:
            episode_dict = {headers[i]:values[i] for i in range(len(values))} #header하나와 values하나와 짝 지어주기
            episodes.append(episode_dict)
for episode in episodes:
    print(episode)
```

```python
import sqlite3
```

```python
query = "CREATE TABLE test(a VARCHAR(20), b VARCHAR(20),c REAL, d INTEGER);" # 20글자를 넘지 않는 a라는 칸 만들기
con = sqlite3.connect('mydata.sqlite')
con.execute(query) # 만들어진 말을 database server에게 전달해야함
```

```python
con.commit() #commit : 데이터베이스에서 영구적인 변경을 확정하는 일(취소가 안됨)
```

```python
data = [('atlanta','georgia','1.25','6'),
       ('Tallahassee','Florida','2.6','3'),
       ('Sacramento','California','1.7','5')]
query = "INSERT INTO test VALUES(?,?,?,?)" # test에다가 VALUES(?,?,?,?)를 추가해라
con.executemany(query,data)
con.commit()
```

```python
query = "SELECT * FROM test" # test라는 표에서부터 모든 것을 다 가져와라
cur = con.execute(query)
rows = cur.fetchall()
rows
```
