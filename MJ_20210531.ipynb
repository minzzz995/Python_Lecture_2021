```python
dataA = np.random.standard_normal((100,2)) #평면에 100개의 점을 찍는다
dataB = np.random.standard_normal((100,2))
dataC = np.random.standard_normal((100,2))
dataA +=np.array((-3,-3))
dataB +=np.array((4,4))
dataC +=np.array((-1,4)) # 각각의 데이터들을 평면이동하였음
plt.scatter(dataA[:,0],dataA[:,1],edgecolor="0",color="b")
plt.scatter(dataB[:,0],dataB[:,1],edgecolor="0",color="0.5") #0~1
plt.scatter(dataC[:,0],dataC[:,1],edgecolor="0",color="#79AAFF") #'색상표' 검색 후 다양한 색상 얻을 수 있음
plt.show()
```

```python
import matplotlib.cm as cm
N = 256
angle = np.linspace(0, 8*2*np.pi,N)
radius = np.linspace(5,1,N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X,Y,c=angle,cmap=cm.Greys)  #cm.hsv / cm.Spectral
plt.show()
```

```python
A = np.random.standard_normal((100,2))
B = np.random.standard_normal((100,2))

A += np.array((-1,-2))
B += np.array((0.5,1))
plt.scatter(A[:,0],A[:,1],color='k',marker='o') #markers종류 : ^, o, O, v, <,P(소문자, 대문자 구분), x, >, X, D....다양함
plt.scatter(B[:,0],B[:,1],color='k',marker='^') # 데이터 숨기러았을때 -> 빈칸
plt.show()
```

```python
X = np.linspace(-6,6,1024)
Y1 = np.sinc(X)
Y2 = np.sinc(X)+1
plt.plot(X,Y1,marker='o',color='0.7')
plt.plot(X,Y2,marker='o',color='0.7',markevery=32) #markevery는 32개의 점마다 하나씩 주기를 줌
plt.show()
```

```python
A = np.random.standard_normal((100,2))
B = np.random.standard_normal((100,2))

A += np.array((-1,-2))
B += np.array((0.5,1))
plt.scatter(A[:,0],A[:,1],s=10) 
plt.scatter(B[:,0],B[:,1],s=100)
plt.show()
```

```python
M = np.random.standard_normal((500,2))
R = np.sum(M**2,axis=1)
plt.figure(figsize=(5,5)) #표 넓이
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.scatter(M[:,0],M[:,1],c='0.7',marker='s',
           s=32*R,edgecolor='k')
plt.show()
```

```python
X = np.linspace(-6,6,1024)
X_small = np.linspace(-3,3,1024)
Y_small = np.sinc(X_small)
plt.plot(X,np.sinc(X),c='0.7')
sub_axes = plt.axes([0.6,0.6,0.25,0.25]) #(0.6,0.6)자리에 0.25만큼 만든다
sub_axes.plot(X_small,Y_small,c='0.0')
plt.show()
```

```python
T = np.linspace(0,2*np.pi,1024)
plt.axes(polar=True) #좌표계를 동그랗게 접는다 -> 일반 좌표계가 아니고 극자표계임
plt.plot(T,1+0.25*np.sin(16*T))
plt.show()
```

```python
import matplotlib.patches as patches
plt.axes(polar=True)
radius = [1,0.5,0.7,0.3,0.1] #,0.2,0.6,0.8,0.5]
theta = np.linspace(0,2*np.pi,len(radius), endpoint=False)
points = np.vstack((theta, radius)).T
plt.gca().add_patch(patches.Polygon(points,color="0.5"))
plt.show()
```
